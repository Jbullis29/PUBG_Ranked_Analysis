```python
import pandas as pd
from config import api_key
import requests
from datetime import date 
import time
```

```python
seasons = [1,2]
game_type = 'solo'
df_list =[]
start_dates= [date(2019, 1, 22),date(2019, 2, 21)]
end_dates = [date (2019, 2, 21),date(2019, 8, 13)]
for x, y, z in zip(seasons, start_dates,end_dates):
    url = f'https://api.pubg.com/shards/pc-na/leaderboards/division.bro.official.xbox-0{x}/{game_type}'
    header= {"Authorization": f"Bearer {api_key}","Accept": "application/vnd.api+json"}
    response = requests.get(url,headers=header)
    season_type = url.split('.')[5]
    season = season_type.split('/')[0]
    type_ = season_type.split('/')[1]
    delta = z-y
    days = delta.days
    json = response.json()
    for attribute in json['included']:
        user_dict = {'name':[], 'account_id':[], 'rank':[],'games':[], 'kills':[],'wins':[],
                 'average_damage':[],'win_ratio':[], 'season':[], 'game_type':[], 'season_length':[]}
        name = f'df_{season}'
        user_dict['name'].append(attribute['attributes']['name'])
        user_dict['account_id'].append(attribute['id'])
        user_dict['rank'].append(attribute['attributes']['rank'])
        user_dict['games'].append(attribute['attributes']['stats']['games'])
        user_dict['kills'].append(attribute['attributes']['stats']['kills'])
        user_dict['wins'].append(attribute['attributes']['stats']['wins'])
        user_dict['average_damage'].append(attribute['attributes']['stats']['averageDamage'])
        user_dict['win_ratio'].append(attribute['attributes']['stats']['winRatio'])
        user_dict['season'].append(f'{season}')
        user_dict['game_type'].append(f'{type_}')
        user_dict['season_length'].append(days)
        name = pd.DataFrame.from_dict(user_dict)
        df_list.append(name)
```

```python
master_rank = pd.concat(df_list)
master_rank.sort_values(by='rank', ascending = True, inplace=True)
master_rank
```

```python
master_rank.set_index('name', inplace=True)
master_rank.to_csv('player_rank.csv')
```

```python
id_list = []
df_list =[]
season_list = []
name_list = []
rank_list =[]
calls = 0
for j,k in master_rank.iterrows():
    calls +=1
    id_ = k['account_id']
    id_list.append(id_)
    name_list.append(j)
    season_list.append(k['season'])
    rank_list.append(k['rank'])
    url = f'https://api.pubg.com/shards/xbox/players/{id_}/seasons/division.bro.official.xbox-01'
    header= {"Authorization": f"Bearer {api_key}","Accept": "application/vnd.api+json"}
    response = requests.get(url,headers=header)
    r = response.json()
    solo_stats = r['data']['attributes']['gameModeStats']['solo']
    id_= pd.json_normalize(solo_stats)
    id_.drop(['boosts','dBNOs', 'dailyKills', 'dailyWins', 'rankPoints', 'rankPointsTitle','revives',
              'weeklyKills','weeklyWins','winPoints','killPoints','teamKills'], axis= 1, inplace=True)
    df_list.append(id_)
    if calls == 10:
        time.sleep(60)
        calls=0
```

```python
top_stats = pd.concat(df_list)
top_stats.insert(0,'name', name_list)
top_stats.insert(1,'account_id', id_list)
top_stats.insert(2,'season', season_list)
top_stats.insert(3,'rank', rank_list)
top_stats
```

```python
top_stats.to_csv('season_one_and_two_ranked_player_stats.csv')
```
