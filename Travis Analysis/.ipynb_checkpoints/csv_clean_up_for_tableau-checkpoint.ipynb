```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("match_samps.csv")
df2 = df.drop(["stats.name", "matchId", "assetsId",
               "matchType", "stats.deathType", "map", 
               "stats.rank", "stats.playerId"], axis=1)
value_count = df2['gameMode'].value_counts()
df_value_counts = pd.DataFrame(value_count)
df_value_counts = df_value_counts.reset_index()
df_value_counts.columns = ['gameMode', 'num_games_played']
df3 = df2[df2['gameMode']=="solo"]
```

```python
df4 = df3.drop(["stats.assists","stats.DBNOs",
                "stats.teamKills","stats.revives",
                "stats.roadKills", "stats.vehicleDestroys",
                "gameMode","stats.swimDistance","stats.killStreaks"],axis=1)
df4.shape
```

```python
plt.figure(figsize=(20,17))
sns.heatmap(df4.corr(), annot=True, cmap="coolwarm").set_title('Seaborn')
plt.savefig('DataTrav/heatmapMatrix.png')
```

```python
pubg_pairplot = df4.dropna()
sns.pairplot(pubg_pairplot, size=1);
```

```python
df4.to_csv('DataTrav/correlation_matrix_clean.xlsx', mode='w', header=False)
```
