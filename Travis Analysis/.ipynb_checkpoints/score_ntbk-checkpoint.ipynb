```python
import pandas as pd
```

```python
file = 'DataTrav/train_test_score.csv'
score_list = pd.read_csv(file)
score_list.reset_index(drop=True, inplace=False)
```

```python
score_list = score_list.drop(score_list.index[[1]])
score_list.reset_index(drop=True, inplace=True)
score_list.columns = ["Rank", "Algorithm", "Test Accuracy", "Training Accuracy"]
score_list.reset_index(drop=True, inplace=True)
del score_list['Rank']
score_list
```

```python
grid_search_row ={'Algorithm':'Grid Search','Test Accuracy':'0.780642', 'Training Accuracy': '0.958899'}
score_list = score_list.append(grid_search_row,ignore_index=True)
score_list
```

```python
score_list.to_csv('DataTrav/score_df2.csv', mode='w', header=False)
```
