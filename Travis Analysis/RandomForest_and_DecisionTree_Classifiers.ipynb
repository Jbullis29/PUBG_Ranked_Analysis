```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Read in data
df = pd.read_csv("match_samps.csv")
df2 = df.drop(["stats.name", "matchId", "assetsId", "matchType", "stats.deathType", "map", "stats.rank", "stats.playerId"], axis=1)
value_count = df2['gameMode'].value_counts()
df_value_counts = pd.DataFrame(value_count)
df_value_counts = df_value_counts.reset_index()
df_value_counts.columns = ['gameMode', 'num_games_played']
df3 = df2[df2['gameMode']=="solo"]
df3 = df3.drop(columns=["gameMode"])
df3
```

# Decision Tree Classifier

```python
from sklearn import tree
```

```python
def f(row):
    if row['stats.winPlace'] <= 5:
        val = 1
    else:
        val = 0

    return val
df3['top_five'] = df3.apply(f, axis=1)
df3
```

```python
target = df3["top_five"]
target_names = ["UnRanked", "Top5"]
data = df3.drop(["top_five", "stats.winPlace"], axis=1)
feature_names = data.columns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
# Score = mean accurracy
```

```python
score_df = pd.DataFrame(columns=["Algorithm", "Accuracy"])
score_df = score_df.append(pd.Series(["Decision Tree Classifier", clf.score(X_test,y_test)], index=["Algorithm","Accuracy"]), ignore_index=True)
score_df.to_csv('DataTrav/score2.csv', mode='a', header=False)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
score_df = score_df.append(pd.Series(["Random Forest Classifier", rf.score(X_test, y_test)], index=["Algorithm", "Accuracy"]), ignore_index=True)
score_df.index = score_df.index + 1
score_df.to_csv('DataTrav/score2.csv', mode='a', header=False)
```
