```python
import pandas as pd
from config import api_key
import requests
from datetime import date 
```

```python
seasons = [1,2]
game_type = 'solo'
df_list =[]
start_dates= [date(2019, 1, 22),date(2019, 2, 21)]
end_dates = [date (2019, 2, 21),date(2019, 8, 13)]
for x, y, z in zip(seasons, start_dates,end_dates):
    url = f'https://api.pubg.com/shards/xbox/leaderboards/division.bro.official.xbox-0{x}/{game_type}'
    header= {"Authorization": f"Bearer {api_key}","Accept": "application/vnd.api+json"}
    response = requests.get(url,headers=header)
    season_type = url.split('.')[5]
    season = season_type.split('/')[0]
    type_ = season_type.split('/')[1]
    delta = z-y
    days = delta.days
    json = response.json()
    for attribute in json['included']:
        user_dict = {'name':[], 'account_id':[], 'rank':[],'games':[], 'kills':[],'wins':[],
                 'average_damage':[],'win_ratio':[], 'season':[], 'game_type':[], 'season_length':[]}
        name = f'df_{season}'
        user_dict['name'].append(attribute['attributes']['name'])
        user_dict['account_id'].append(attribute['id'])
        user_dict['rank'].append(attribute['attributes']['rank'])
        user_dict['games'].append(attribute['attributes']['stats']['games'])
        user_dict['kills'].append(attribute['attributes']['stats']['kills'])
        user_dict['wins'].append(attribute['attributes']['stats']['wins'])
        user_dict['average_damage'].append(attribute['attributes']['stats']['averageDamage'])
        user_dict['win_ratio'].append(attribute['attributes']['stats']['winRatio'])
        user_dict['season'].append(f'{season}')
        user_dict['game_type'].append(f'{type_}')
        user_dict['season_length'].append(days)
        name = pd.DataFrame.from_dict(user_dict)
        df_list.append(name)
```

```python
master_rank = pd.concat(df_list)
master_rank.sort_values(by='rank', ascending = True, inplace=True)
```

```python
master_rank.reset_index(inplace=True)
```

```python
master_rank.drop('index', axis =1, inplace=True)
master_rank.to_csv('player_rank.csv')
```
